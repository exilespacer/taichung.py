# 階層式分群法 Hierarchy Cluster Analysis
* 一種「分群」的技術，將距離近的資料點分為同一群，得到一棵 Dendrogram (二元樹圖)<img src='fig/dendrogram.png' height=600>

### 分群方式
* divisive  （自上而下）：從樹頂開始分裂，逐層往下分群
* agglomerative （自上而下）：從樹底開始聚合，逐層往上分離
    <br>每個資料點屬於一個群，計算群之間的距離，將距離最近的兩個群合併，重新計算群的距離以及合併，重複至只剩一個群為止

### 定義「距離」
#### 點距離
* 歐式距離
* 相關性距離

### 定義「距離」
#### 群距離: linkage
    * single-linkage: 最近
    * complete-linkage：最遠
    * group average：兩組間各點的平均距離
    * ward's method: 合併後到群中心的距離

<img src='fig/linkage.jpg'>

### 小心使用
* 優點：
    * 概念簡單，只需要知道兩兩距離
    * 事先不需要知道群數，依據 Dendrogram 數可決定橫切點，得到指定的群數
 
* 缺點：只適用於少量資料
