```python
import pandas as pd
import matplotlib.pylab as plt
import seaborn as sns
```

```python
df_all = pd.read_csv('data/factor_grps.csv')
df = df_all.query('factor=="factor0"')
df
```

```python
pd.ols(y=df.measure, x=df.group)
```

```python
sns.lmplot("group", "measure", df) 
```

```python
def get_factor_regression_line(df, factor_id):
    factor = 'factor'+str(factor_id)
    df = df_all.query('factor=="%s"'%factor)
    sns.lmplot("group", "measure", df) 
```

```python
for i in range(10):
    get_factor_regression_line(df_all, i)
```
